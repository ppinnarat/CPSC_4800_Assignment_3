```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
train = pd.read_csv("../Data/train.csv")
test = pd.read_csv("../Data/test.csv")
survived = pd.read_csv("../Data/gender_submission.csv")
```

```python
train
```

```python
test
```

```python
survived
```

```python
test1 = pd.merge(survived,test,how='inner',on='PassengerId')
test1
```

```python
titanic = pd.concat([train,test1])
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.isnull().sum()
```

```python
titanic_cleaned = titanic.fillna(value = titanic["Age"].mean())
titanic_cleaned
```

Hypothesis 1: The survival rate is associated to the class of passenger

```python
survival_by_class = titanic.groupby("Pclass")["Survived"].mean()
survival_by_class
```

```python
sns.barplot(x = "Pclass", y = "Survived", data = titanic, color = "lightblue") 
plt.title("The bar chart of the survival rate and the class of passenger")
plt.xlabel("The class of passenger")
plt.ylabel("The survival rate")
plt.savefig("../Result/The bar chart of the survival rate and the class of passenger.png")
plt.show()
```

Hypothesis 2: The survival rate is associated to the gender

```python
survival_by_gender = titanic.groupby("Sex")["Survived"].mean()
survival_by_gender
```

```python
sns.barplot(x = "Sex", y = "Survived", data = titanic, color = "lightgreen") 
plt.title("The bar chart of the survival rate and gender of passenger")
plt.xlabel("The gender of passenger")
plt.ylabel("The survival rate")
plt.savefig("../Result/The bar chart of the survival rate and gender of passenger.png")
plt.show()
```

Hypothesis 3: The survival rate is associated to the age

```python
correlation = titanic[["Survived", "Age"]].corr()
correlation
```

```python
sns.heatmap(correlation)
plt.savefig("../Result/The heatmap.png")
plt.show()
```

```python
sns.histplot(titanic["Age"], bins = 20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.savefig("../Result/The histogram of age.png")
plt.show()
```
