```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
test = pd.read_csv("../Data/test_house.csv")
train = pd.read_csv("../Data/train_house.csv")
sale_price = pd.read_csv("../Data/sample_submission.csv")
```

```python
train
```

```python
test
```

```python
sale_price
```

```python
test1 = pd.merge(test,sale_price,how='inner',on='Id')
test1
```

```python
house = pd.concat([train, test1])
house
```

```python
house.info()
```

```python
house.describe()
```

```python
house.isnull().sum()
```

```python
missing_percentage = house.isnull().mean()
cols_to_drop = missing_percentage[missing_percentage > 0.4].index
house_cleaned = house.drop(columns=cols_to_drop)
num_cols = house_cleaned.select_dtypes(include=['number']).columns
house_cleaned[num_cols] = house_cleaned[num_cols].fillna(house_cleaned[num_cols].median())
house_cleaned[num_cols]
```

```python
cat_cols = house_cleaned.select_dtypes(include=['object']).columns
house_cleaned[cat_cols] = house_cleaned[cat_cols].fillna(house_cleaned[cat_cols].mode().iloc[0])
house_cleaned[cat_cols]
```

```python
house_cleaned = house_cleaned.drop_duplicates()
house_cleaned.info(), house_cleaned.isnull().sum().sum()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(house['SalePrice'], bins=30, kde=True, color='blue')
plt.title('Distribution of Sale Price', fontsize=14)
plt.xlabel('Sale Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.savefig("../Result/Distribution of Sale Price.png")
plt.show()
```

```python
correlation = house.corr(numeric_only=True)
correlation
```

```python
top10_correlations = correlation["SalePrice"].sort_values(ascending=False).head(10)
top10_correlations
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation[top10_correlations.index].loc[top10_correlations.index], 
            annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Top Correlated Features with SalePrice", fontsize=14)
plt.savefig("../Result/Top Correlated Features with SalePrice.png")
plt.show()
```
